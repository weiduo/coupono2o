```python
#处理数据，获得分类指标
#oneline
import re
offline_train = offline_train[(True-offline_train['Discount_rate'].isin(['null']))]

def convert_null(x):
    if x == 'null':
        return -1
    else:
        return x
offline_train['Distance'] = offline_train['Distance'].map(convert_null)

Date_received = pd.to_datetime(offline_train.Date_received,format='%Y%m%d',errors='coerce')
Date = pd.to_datetime(offline_train.Date,format='%Y%m%d',errors='coerce')
Time_difference = Date-Date_received
days_15 = pd.Timedelta('15 days')

def isin_15(timedelta):
    if timedelta <= days_15:
        return 1
    else:
        return 0
    
offline_outcome = Time_difference.map(isin_15)
offline_outcome.value_counts()
print len(offline_outcome)
```

```python
from sklearn import preprocessing
#去掉无意义的Coupon_id
offline_train = offline_train.drop(['Coupon_id','Date','Date_received'],axis = 1)
#offline_train = offline_train[(True-offline_train['Discount_rate'].isin(['null']))]

# TODO：使用pandas.get_dummies()对'features_raw'数据进行独热编码
#features_train = pd.get_dummies(offline_train) 
encoded_train = list(offline_train.columns)
#print "{} total features after one-hot encoding.".format(len(encoded_train))

# 移除下面一行的注释以观察编码的特征名字
print encoded_train
display(offline_train.head(n=10))
print offline_train.Distance.value_counts()
print offline_train.Discount_rate.value_counts()
print offline_test.Distance.value_counts()
print offline_test.Merchant_id.value_counts()
```

```python
def splita(x):
    if ":" in x:
        a = x.split(":")
        return float(a[0])/float(a[1])
    else:
        return float (x)

offline_train['Discount_rate'] = offline_train['Discount_rate'].map(splita)
```

```python
# 导入 train_test_split
from sklearn.model_selection import train_test_split

# 将'features'和'income'数据切分成训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(offline_train, offline_outcome, test_size = 0.2, random_state = 0)

# 显示切分的结果
print "Training set has {} samples.".format(X_train.shape[0])
print "Testing set has {} samples.".format(X_test.shape[0])
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, recall_score

#随机设outcome为1，求准确率
offline_outcome_pred = []
for i in range(1053282):
    offline_outcome_pred.append(1)
    
# TODO： 计算准确率
accuracy = accuracy_score(offline_outcome, offline_outcome_pred)
recall = recall_score(offline_outcome, offline_outcome_pred, average='macro')  

# TODO： 使用上面的公式，并设置beta=0.5计算F-score
fscore = (1+0.5**2)*((accuracy*recall)/(0.5**2*accuracy+recall))

# 打印结果
print "Naive Predictor: [Accuracy score: {:.4f}, F-score: {:.4f}]".format(accuracy, fscore)
```
